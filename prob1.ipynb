```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Weather Dataset.csv")
```

```python
df = df.dropna()
df = df.drop_duplicates()
df = df.dropna(axis=1)
df = df[(df['Temperature (C)'] >= -50) & (df['Temperature (C)'] <= 50)]
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df.sort_values(by='Formatted Date', inplace=True)
plt.figure(figsize=(8, 6))
plt.plot(df['Formatted Date'], df['Temperature (C)'],label = 'Temperature (C)')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.title('Temperature-time graph')
plt.xticks(df['Formatted Date'][::200],  rotation=90)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Temperature (C)'], bins=10 ,color='orange' ,edgecolor='black')
plt.xlabel('Temperature (C)')
plt.title('Temperature Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Temperature (C)'], df['Humidity'])
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity')
plt.axis([0,30,0,1])
plt.show()
```

```python
# Generate a correlation heatmap of the dataset
plt.figure(figsize=(8, 6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.2)
plt.title('Correlation Heatmap of Weather')
plt.show()
```
