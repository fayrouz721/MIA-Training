SAMPLING

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2)
y = -1.4 * (x - 1)**2 + 1.4

vlines = np.arange(0, 2.25, 0.25)
for v in vlines:
    plt.plot([v, v], [0, -1.4 * (v - 1)**2 + 1.4], color='red', linewidth=0.8)

plt.plot(x, y, label="f(x)")
plt.title("Plotting Example")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.legend()
plt.show()
```

Quantization

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2)
y = -1.4 * (x - 1)**2 + 1.4

points_x = [0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]
points_y = [0, 0.4, 1, 1.2, 1.4, 1.2, 1, 0.4, 0]

vlines = np.arange(0, 2.25, 0.25)
for v in vlines:
    plt.plot([v, v], [0, -1.4 * (v - 1)**2 + 1.4], color='red', linewidth=0.8)
hlines = np.arange(0, 1.5, 0.2)  
for h in hlines:
   plt.plot([0, 2], [h, h], color='green', linewidth=0.8,linestyle = "dashed")
plt.plot(points_x, points_y, 'bo')
plt.plot(x, y, label="f(x)")
plt.title("Doom Signal")
plt.xlabel("Time (seconds)")
plt.ylabel("Voltage")
plt.legend()
plt.show()
```

Discrete Signal

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2)
y = -1.4 * (x - 1)**2 + 1.4

points_x = [0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2]
points_y = [0, 0.4, 1, 1.2, 1.4, 1.2, 1, 0.4, 0]

binary_values = ['000', '001', '010', '011', '100', '101', '110', '111']
y_values = np.linspace(0, 1.4, len(binary_values))  
fig, ax1 = plt.subplots()

for i in range(len(points_x)):
    px = points_x[i]
    py = points_y[i]
    ax1.plot([px, px], [0, py], color='blue', linewidth=2.5)

hlines = np.arange(0, 1.5, 0.2)  
for h in hlines:
    ax1.plot([0, 2], [h, h], color='green', linewidth=0.8, linestyle="dashed")

ax1.plot(points_x, points_y, 'bo')
ax1.set_xlabel("Time (seconds)")
ax1.set_ylabel("Voltage")
ax1.set_title("Doom Signal")
plt.plot(x, y)
ax2 = ax1.twinx()  
ax2.set_ylim(ax1.get_ylim())  
ax2.set_yticks(y_values)
ax2.set_yticklabels(binary_values)
ax2.spines['right']
ax2.yaxis.tick_left()  
ax2.spines['left'].set_position(('outward', 60)) 
plt.show()
```
